```python
df = pd.read_csv("F:/Data_Files/Twitter_Data/Election2016/Unlabeled_Data/Unlabeled_Trump_ElectionDay.csv")
```

```python
del df['Unnamed: 0']
del df['User_ID']
del df['Tweet_ID']
del df['sentiment']
del df['multi-team']
```

```python
df['Tweet'] = df['Tweet'].str.replace('#','')
df['Tweet'] = df['Tweet'].str.replace('@','')
```

```python
df.tail(25)
```

```python
df.Tweet = df.Tweet.astype(str)

import textblob
from textblob import TextBlob

# The x in the lambda function is a row (because I set axis=1)
# Apply iterates the function accross the dataframe's rows
df['polarity'] = df.apply(lambda x: TextBlob(x['Tweet']).sentiment.polarity, axis=1)
df['subjectivity'] = df.apply(lambda x: TextBlob(x['Tweet']).sentiment.subjectivity, axis=1)
```

```python
df.tail()
```

```python
def sent_pred(c):
    if c['polarity'] > 0:
        return 'positive'
    elif c['polarity']  < 0:
        return 'negative'
    elif c['polarity']  == 0:
        return 'neutral'
    else:
        pass        

df['Sentiment_Prediction'] = df.apply(sent_pred, axis=1)
```
