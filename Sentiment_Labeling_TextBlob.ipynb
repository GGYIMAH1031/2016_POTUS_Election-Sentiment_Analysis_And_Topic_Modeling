```python
df = pd.DataFrame()
```

```python
df['Country'] = Train_data['Country']
df['tweet'] = Train_data['Tweet']
```

```python
df['tweet'] = df['tweet'].str.replace('#','')
df['tweet'] = df['tweet'].str.replace('@','')
```

```python
df.tail(25)
```

```python
df.tweet = df.tweet.astype(str)

import textblob
from textblob import TextBlob

# The x in the lambda function is a row (because I set axis=1)
# Apply iterates the function accross the dataframe's rows
df['polarity'] = df.apply(lambda x: TextBlob(x['tweet']).sentiment.polarity, axis=1)
df['subjectivity'] = df.apply(lambda x: TextBlob(x['tweet']).sentiment.subjectivity, axis=1)
```

```python
df.tail()
```

```python
def sent_pred(c):
    if c['polarity'] > 0:
        return 'positive'
    elif c['polarity']  < 0:
        return 'negative'
    elif c['polarity']  == 0:
        return 'neutral'
    else:
        pass        

df['Sentiment_Prediction'] = df.apply(sent_pred, axis=1)
```
